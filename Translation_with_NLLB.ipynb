```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
translator_dutch = pipeline('translation', model=model, tokenizer=tokenizer, src_lang="tur_Latn", tgt_lang='nld_Latn', max_length=400)
translator_dutch("selam güzel çevirebildim mi")
```

```python
translator_german = pipeline('translation', model=model, tokenizer=tokenizer, src_lang="tur_Latn", tgt_lang='deu_Latn', max_length=400)
translator_german("selam güzel çevirebildim mi?")
```
